# Coral Bleaching: Getting Worse? 
Coral bleaching can occur when temperatures [exceed a certain threshold that a reef can withstand](https://coralreefwatch.noaa.gov/satellite/methodology/methodology.php#:~:text=The%20HotSpot%20value%20of%201.0,and%20are%20updated%20twice%2Dweekly.). When corals become heat stressed, they expel their zooxanthellae with whom they have a mutualistic reltaionship and depend on for survival. Corals are usually able to cope with heat unless temperatures go above a certain threshold, usually 1 degree C above the normal maximum temperature for a given month.     
In our study area of Southern Tonga, this threshold is [28.8 degrees Celsius](https://coralreefwatch.noaa.gov/product/vs/gauges/southern_tonga.php).    
Using a [NOAA dataset](https://coralreefwatch.noaa.gov/product/vs/timeseries/polynesia.php#southern_tonga) of satellite sea surface temperatures (SST) dating back to 1985, we can visualize and quantify how, if at all, bleaching events have changed in intensity or frequency over the past several decades. SST is just a proxy for bleaching events — sometimes heat waves are brief enough not to cause bleaching, sometimes other environmental stressors make corals more vulnerable hot temperatures. Regardless, finding trends in the occurrence of "bleaching conditions" can help us understand current and future threats to reefs.

### Step 1: Import Data from NOAA   
NOAA's SST datasets have a lot of info we won't need to use. We will make sure we have correct date info as well as daily maximum temperatures, which are under the "SST_MAX" heading:

```python
#Import modules
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import axes as ax
import matplotlib.cm as cmap
from matplotlib import colors
import pandas as pd
```

```python
#Import temperature data
csv = 'southern_tonga_timeseries.csv'
data = pd.read_csv(csv)
data
```

### Step 2: Quick plot of the data   
In our dataset, each row contains temperature data for one day. NOAA has a somewhat complicated methodology for determining what temperature patterns trigger different "coral bleaching alert levels", which you can read more about [here](https://coralreefwatch.noaa.gov/satellite/methodology/methodology.php#:~:text=The%20HotSpot%20value%20of%201.0,and%20are%20updated%20twice%2Dweekly.).    
For now, let's visualize daily maximum SST over time in Tonga, with an eye for temperatures that occur above the bleaching threshold. We'll plot a dashed red line at the bleaching threshold of 28.8 degrees C:

```python
#Simple plot of temp data over time
plt.figure(figsize=(15,10))
plt.plot('DAY','SST_MAX',data=data)
plt.axhline(y=28.8, color='r', linestyle='--')
plt.title('Max Daily SST, 1985-present')
plt.xlabel('Year')
plt.ylabel('Temperature (deg C)')
#Fix tick labels
tickloc = np.arange(1,len(data['DAY']),step=(365*5))
ticklab = np.arange(1985, 2022, step=5)
plt.xticks(tickloc,ticklab)
```

### Step 3: Analyzing trends   
Great, we can see our daily max temperature plotted over the period of record. We can also see that temperatures have gone over the bleaching threshold A LOT over the period of record. It's a little hard to see how severe these heat events were, though, or if there are any increases in frequency or severity over time.   
We can visualize how the number of bleaching days has changed year to year. Let's filter out days with max daily temperatures above the bleaching threshold and count the number of bleaching days per year. We can then visualize with a simple linear regression:

```python
#filter dataset by days with bleaching temperatures, count bleaching days by year
bleach_days = data[data.SST_MAX >= 28.8]
bleach_days_by_year = bleach_days['YYYY'].value_counts(sort=False)
#plot the data
plt.figure(figsize=(15,10))
bleach_days_by_year.plot(kind='line',color='gray')
#calculate linear regression for trendline
x = np.arange(1985,2023,step=1)
z = np.polyfit(x, bleach_days_by_year, 1)
p = np.poly1d(z)
#add trendline, labels to plot
plt.plot(x, p(x),color='r')
plt.title('Number of Bleaching Days Per Year, Southern Tonga')
plt.xlabel('Year', fontsize=18)
plt.ylabel('Bleaching Days (Max SST above 28.8 deg C) per Year', fontsize=16)
```

### Discussion
The number of bleaching days per year has most certainly increased since 1985. Although the conditions that cause coral mortality in Southern Tonga is a slightly more complicated picture than simply the number of days with SST above 28.8 C, heat stress on corals appears to have gotten much worse in recent decades.
